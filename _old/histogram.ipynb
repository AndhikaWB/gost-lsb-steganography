```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
cover = input("Masukkan path ke gambar cover: ")
print("Gambar cover:", cover)

# stego = input("Masukkan path ke gambar stego: ")
stego = cover.rsplit(".", 1)
stego = stego[0] + "_stego." + stego[1]

# Konversi gambar ke grayscale
cover = Image.open(cover).convert("L", colors = 256)
stego = Image.open(stego).convert("L", colors = 256)

print("Ukuran gambar:", cover.size)
```

```python
# Buka gambar sebagai array dan konversi ke 1 dimensi
cover_array = np.asarray(cover).ravel()
stego_array = np.asarray(stego).ravel()
```

```python
fig = plt.figure(figsize = (10, 5))

fig.add_subplot(1, 2, 1)
plt.hist(x = cover_array, bins = 256, range = [0, 256], color = "blue")
plt.title("Gambar Cover (Grayscale)")
plt.ylabel("Intensitas Piksel")
plt.xlabel("Rentang Nilai Piksel")

fig.add_subplot(1, 2, 2)
plt.hist(x = stego_array, bins = 256, range = [0, 256], color = "red")
plt.title("Gambar Stego (Grayscale)")
plt.ylabel("Intensitas Piksel")
plt.xlabel("Rentang Nilai Piksel")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(x = cover_array, bins = 256, range = [0, 256], color = "blue", label = "Gambar Cover")
ax.hist(x = stego_array, bins = 256, range = [0, 256], color = "red", label = "Gambar Stego")
ax.set_title("Histogram Gambar (Grayscale)")
ax.set_xlabel("Rentang Nilai Piksel")
ax.set_ylabel("Intensitas Piksel")
plt.legend()
plt.show()

# Zoom histogram untuk melihat perbedaan
# Pastikan format grafik matplotlib dalam vector
# Gunakan "jupyter.generateSVGPlots = true" pada VS Code
```
